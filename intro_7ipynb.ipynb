## Richiami sulla distribuzione normale univariata e multivariata

Richiami - Normale univariata
- in ambito univariato , la variabile casuale v.c. quantitativa continua X ha lege di distribuzione normale se la sua funzione di densita di probabilita f.d. e' data da:

$f(x;\mu;\sigma_2)= \frac{1}{\sqrt{2\pi\sigma_2}}exp[-\frac{1}{2\sigma^2}(x-\mu)^2]$

con: $x \in R, mu \in R,\sigma>0$. Si verifica che: $E[X]= \mu, Var[X]=\sigma^2$

- siscrive $X\sim N(\mu;\sigma^2)$

La v.c. p -dimensionale $X= (x_1...X_p), p \geq2$ quantitativa continua ha legge di distribuzione normale multivariato di parametri $\mu$ e $\sum$ (e si scrive $X \sim N_p (\mu, \sum)$) se ;a sua f.d. ha la firma

$f(x\mu, \sum) = \frac{1}{(2r)^{p/2} |\sum|^{1/2}}.exp[-\frac{1}{2} (x-\mu)^t \sum^{-1}(x-\mu)]$

dove $x \in R^p$ e' il generico vettore di valori di X; $\mu=[\mu_1...\mu_p]^t \in R^p $ e' il vettore dei valori attesi di X, ossia

$\mu=E[X] = [E(X_1), E(X_2)...E[X+p]]^t= [\mu_1...\mu_p]^t$

### Proprieta per transformazioni lineari di una v.c. MV quantitativa:
- transformazione lineare scalare:
- $V = 1+b^tX$ dove $a \in R$ e $b=[b_1...b_p]^t \in R^p$

- con almeno un $b_j \ne 0$ o anche $b \ne 0$

Allora:
- valore attesso di V: $\mu_v = E[V]= a + b^tE[X]= a+b^t\mu_x$

- varianza di V: $Var[V] = Var(a + b^tX)= b^t \sum b$

Proprieta per transformazioni lineari di una v.c. MV quantitatva:
- transformazione lineare vettoriale:
$V_{q*1}= a_{1*1} + B_{q*p} X_{p*1}$ dove $a=[a_1...a_q]^t \in R^q$ e $B$$ , matrice ad elementi non tutti nulli $B \ne 0_{\text{matrice nulla}}$

- vettpre dei valori atesi: $\mu_v = E[V]= a + BE[X]= a + B\mu_x$

- matrice di var-cov di V:
- $V(V)_{q*q} = B_{q*p} V[X]_{p*p} B^t = B\sumB^t$

### Proprieta di indipendenza stocastica:
- in generale data la v.c p-variata quantitativa $X={X_1...X_p}$ con f.d $f(x)$ si dice che le sue componenti $X_j$ sonoIndipoendenti in senso distributivo o staocastico se:

$f(x)= f(x_1)...f(x_j)..f(x_p)= \prod_{j=1}^p f(x_j)$

dove: $f(x)$ e' la f.d. congiunta e $f(x_j)$ spmp f.d marginali dell v.c. $X_j$


### Proprieta di indipendenza stocastica
- indipendenza stocastica I.S. come forma piu forte di indipendenza . In generale si dimostra:
