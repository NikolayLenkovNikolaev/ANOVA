## 1. Intervallidi confidenza per i parametri del modello
- metodo delle quantita pivotale
- risultato teorico di partenza:


$\hat{B_j} \sim N(\beta_j; \sigma^2 c_{jj}) => Z_j = \frac{\hat{B_j}- \beta_j}{\sqrt{\sigma^2 c_{jj}}} \sim N(0,1)$

v.c nmormale standardizzata - quantita pivotale

Problema: $\sigma^2$ e' ignota e oggetto di stima -> si ricorre allo stimaore $f^2$ - varianza vompionaria corretta


$\frac{\hat{B_j}- \beta_j}{\sqrt{f^2 c_{jj}}} = \frac{\frac{\hat{B_j}- \beta_j}{\sqrt{\sigma^2 c_{jj}}}}{\sqrt{\frac{f^2 c_{jj}}{\sigma^2 c_{jj}}}}$


$= \frac{N(0,1)}{\sqrt{\frac{\chi_{n-(k+1)}^2}{n-(k+1)}}} \sim T_{n-(k+1)}$
- $\sim$ qua $\hat{B_j}$ and $f^2$ IS
- $T_{n-(k+1)}$- v.c t di Student con (n-(K+1) gfl)


$=> \frac{f^2}{\sigma^2} \sim \frac{\chi_{n-(k+1)}^2} {n-(k+1)}$

Costruzione dello Stimatore Intervallare (intervalllo casuale) per $\beta_j$ (j=0,1...K) al livello di confidenza $1-\alpha; 0<\alpha <1$

$P[-t_{\alpha/2} \leq T_{n-(k+1)} \leq t_{\alpha/2}] = 1-\alpha $
- dove $-+ t_{\alpha/2}$ son valori di riferimento della vc. $T_{n-(k+1)}$ individuati con $1-\alpha$ prefissato

$P[   -\hat{B_j} -t_{\alpha/2} \sqrt{f^2 c_{jj}} \leq -\hat{\beta_j} \leq  -\hat{B_j}  + t_{\alpha/2} \sqrt{f^2 c_{jj}} ] = 1-\alpha$

$P[  \hat{B_j} -t_{\alpha/2} \sqrt{f^2 c_{jj}} \leq \hat{\beta_j} \leq  \hat{B_j}  + t_{\alpha/2} \sqrt{f^2 c_{jj}} ] = 1-\alpha$

dove:
- $P[T_{n-(k+1)} \geq t_{\alpha/2}] = P[T_{n-(k+1)} \leq t_{\alpha/2} ] =\frac{\alpha}{2}$

Stimare intervallo al livello di confidenza $1-\alpha$

$[\hat{B_j}- t_{\alpha/2} \sqrt{f^2 c_jj}; \hat{B_j} + t_{\alpha/2} \sqrt{f^2 c_jj}]$

Sul singolo campione: intervallo di confidenza- CI -confidence interval

$[\hat{B_j}- t_{\alpha/2} \sqrt{s^2 c_jj}; \hat{B_j} + t_{\alpha/2} \sqrt{s^2 c_jj}]$

- $\hat{\beta_j} -+ t_{\alpha/2} \sqrt{s^2 c_jj}$

- $\hat{\beta_j} -+ t_{\alpha/2} S.E (\hat{\beta_j})$

dove: $S.E(\hat{\beta_j}) = \sqrt{s^2 c_{jj}}$
- S.E/ Standard Error delo stimatore $\hat{B_j}$- ottenuto dal campione



## 2. Verifica di ipotesi statistiche v.d.i

### a. Test sull'intera regressione|: test F

$H_0: \beta_1=\beta_2=...=\beta_j = ... =\beta_k =0$ contro $H_1: \beta_j \ne 0$ per almeno un j

NB: $\beta_0$e' esclusa da queta v.d.i per ragioni  teoriche


v.d.i alla base della tabella ANOVA
- costruzione della statistica Test

- leggi di distribuzione relative alle v.c. "Devianze"

1. V.c Devianza Residua

$\frac{\sum_{i=1}^n (Y_i - \hat{Y_i}^2)}{\sigma^2} \sim \chi_{n-(k+1)^2}$
- vale sempre sotto le ipotesi da a a d

- $\sum_{i=1}^n (Y_i - \hat{Y_i}^2)$ - SSE

2. V.c Devianza SPiegata

$\frac{\sum_{i=1}^n (Y_i - \bar{Y_i}^2)}{\sigma^2} \sim \chi_{k)^2}$
- sotto $H_0: \beta_1..=\beta_k=0$ - oltre alla validita delle ipotesi da a-d

3. V.c devianza totale (per completezza)

$\frac{\sum_{i=1}^n (Y_i - \bar{Y_i}^2)}{\sigma^2} \sim \chi_{n-1)^2}$
- sotto $H_0: \beta_1..=\beta_k=0$ - oltre alla validita delle ipotesi da a-d


Inoltre: v.c SSE e v.c SSR IS

Statistica test:

Teorema: legame fra v.c Chi-quaddrato e vc F di Fisher

Date due v.c $\chi_{g1}^2$ e$\chi_{g_2}^2$ con $q_1,g_2$ gdl ripetivamente e fra loro IS. si ha:

$$\frac{\frac{ \chi_{g_1}^2 }{g_1 }}{\frac{\chi_{g_2}  }{g_2  }} \sim \mathcal{F{g_1,g_2}}$$


Nel caso della v.d.i sull intera ragressione

$$ \frac{\frac{\sum_{i=1}^n (\hat{Y_i -\bar{Y}})^2   }{ k\sigma^2   }}{\frac{ \sum_{i=1}^n (Y_i - \hat{Y_i})^2  }{ (n-k-1)\sigma^2   }} =  \frac{\frac{ \chi_{k}^2 }{K }}{\frac{\chi_{n-(k+1)}  }{n-(k+1)  }}$$

$\frac{\frac{MSR}{\sigma^2}}{\frac{MSE}{\sigma^2}} \sim \mathcal{F}_{k, n-(k+1)}$

dove $\mathcal{F_{n, n-(k+1)}}$ e' la v.c F di Fisher con k-gdl al numeratore e $n-(k-1)$ gdl al denominatore

Svolgimento dell vdi sul campione
- si calcola il valore empirico della statistica test $\mathcal{F_{n, n-(k+1)}}$

$F_e = \frac{\frac{SSR}{K}}{\frac{SSE}{n-(k+1)}} = \frac{MSR}{MSE}$

- si individui il valore critico $F_{\alpha}$ di riferimento per il test sulla distribuzione della statistica test $\mathcal{F_{n, n-(k+1)}}$ in corrispondenza del Livelo di Significativita (NOMINALE)

prefissato pari a $\alpha; 0 < \alpha < 1$

$P[\mathcal{F_{n, n-(k+1)}} \geq F_{\alpha}] =\alpha$

image.png

Regola decisionale : dal confronto fra $F_e$ e $F_{\alpha}$ si stabilisce se accettre o rifiutare $H_0$

In particolare, se:
- $F_e < F_{\alpha}$ - si accetta $H_0$
- $F_2 \geq F_{\alpha}$ - si rifiuta $H_0$

In altrnativa si opera un confronto fra il livello di significativita $\alpha$ e il livello di significativita empirico , detto p-value

p-value = $p[\mathcal{F_{n, n-(k+1)}} \geq F_2]$

Quindi , la regola decisionale e':

- p-value > $\alpha$ - si accetta $H_0$
- p_vlaue $\leq \alpha$ - si rifiuta $H_0$

## b. Tesrt sui singoli parametri; test t

$H_0: \beta_j=0 \text{ vs } H_1: \beta_j \ne 0; j=0,1..k$

Statistica test:

$\frac{\hat{B_j}- \beta_j}{\sqrt{\sigma^2 c_{jj}}} = \frac{\hat{B_j}- }{\sqrt{\sigma^2 c_{jj}}} \sim T_{n-(k+1)}$
- soto $\beta_j=0$

- sul campione: valere empirico $t_e$

$t_e = \frac{\hat{\beta_j}}{S.E (\hat{\beta_j})}; j=0,1...K$

- $SE(\hat{\beta_j}) =\sqrt{s^2 c_jj}$ -all j

Regola decisionale: fissato il livello di significativita $\alpha$ si ha:




image.png

$P[ T_{n-(k+1)} \geq |t_{\alpha/2}|] =\alpha$

Opurre: $P[ T_{n-(k+1)} \geq t_{\alpha/2}] = \frac{\alpha}{2}]$ and $P[ T_{n-(k+1)} \leq - t_{\alpha/2}] = \frac{\alpha}{2}]$

In modo equivalente, tenendo presente che:

$P[ T_{n-(k+1)} \geq |t_e|]$ = p-value

$P[ T_{n-(k+1)} \geq t_{\alpha/2}]=\alpha$

la regole decisonale descrive:
- p-value > $\alpha$ oppure $\frac{p-value}{2} > \frac{\alpha}{2} $ si accetta $H_0$

- p-value $\leq \alpha$ oppure $\frac{p-value}{2} \leq \frac{\alpha}{2} $ si rifiuta $H_0$

Osservazione: Nel caso della retta: K=1

$\mathcal{F_{1, n-2}} = T_{n-2}^2$

test F e test t equivalente ai fini della v.di: $H_0: \beta=0$ vs $H_1: \beta_1 \ne -$
