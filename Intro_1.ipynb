## Coefficiente di Correlazione Lineare di X e Y

$Cor(X,Y)= \rho(X,Y)=\rho_{xy}= \frac{Cov(X,Y)}{\sqrt{Var[X]Var[y]}}= \frac{\sigma_{xy}}{\sigma_x.\sigma_y}$


Formula indiretta:

$Covariance(X,Y)= \frac{1}{N} \sum_{i=1}^N (x_i-\mu_x)(y_i-\mu_y)$

$\frac{1}{N} \sum_{i=1}^N x_iy_i - \mu_x.\mu_y = M(XY)- M(X)M(Y)= \mu_{xy}- \mu_x.\,u_y$

$Cov(X,Y) \in R: Cov \in (-\infty; +\infty)$


Disegnaglianza di Cauchi-Shwartz:

$ - \sqrt{Var[X]Var[Y]} \leq Cor (X,Y)\leq + \sqrt{Var[X]Var[Y]} $

$Cov(X,Y) \in [-\sigma_x\sigma_y ; +\sigma_x\sigma_y]$

$\rho_{xy}= \frac{\sigma_{xy}}{\sigma_x\sigma_y} \in [-1;+1]$

- $\rho=-1$
- $\rho=0$
- $\rho=+1$

Esempi relativi ai valori estremi del range di variazione di $\rho$

- $\rho=- 1 <=> \sigma_{xy}= -\sigma_x\sigma_y$
- $\rho= + 1 <=> \sigma_{xy}= +\sigma_x\sigma_y$

- $\rho = 0$

## Criteri dei minimi quadrati - ordinari - MQ

OLS- Ordinary Least Squares:

**Def di criterio OLS: **

$(x_i, y_i); i=1...N$

$D(\beta_0, \beta_1...\beta_k) = \sum_{i=1}^N [y_1 -f(x_i, \beta_0, \beta_1...\beta_k)]^2 = minimo$

$D(\beta_0, \beta_1...\beta_k) = \sum_{i=1}^N (y_i - \hat{y_1})^2 = \sum_{i=1}^N \epsilon_i^2$

Vettore dei parametri ignoti:

$\beta_{(k+1)*1} = [\beta_0, \beta_1...\beta_k]^t$

Vettore dei valori osservtri di Y - observed values

$y_{N*1} = [y_1...y_i...y_n]^t$

Bettore dei valori presenti di Y sulla base della funzione interpolante - fitted values

$\hat{y_{N*1}} = [\hat{y_1}...\hat{y_i},...\hat{y_N}]^t$

Vettore dei residui o errori
- $\epsilon_{N*1} = [\epsilon_1,...\epsilon_i ...\epsilon_N]^t$



$y - \hat{y} =  [y_1...y_i...y_n]^t -  [\hat{y_1}...\hat{y_i}...\hat{y_N}]^t$

$[y_1 - \hat{y_1}, ....y_1-\hat{y_i} ...y_N - \hat{y_N}]^t = [\epsilon_1,...\epsilon_i ...\epsilon_N]^t = \epsilon$

Prodotti righe per colonne:

$\sum_{i=1}^N (y_1 - \hat{y_i}) = [y_1 - \hat{y_1}, ....y_1-\hat{y_i} ...y_N - \hat{y_N}] . [y_1 - \hat{y_1}, ....y_1-\hat{y_i} ...y_N - \hat{y_N}]^t$

$(y_1-\hat{y_1}^2) +...+(y_i-\hat{y_i}^2)+..+(y_N-\hat{y_N}^2)$

$(y-\hat{y})^t (y-\hat{y})= \epsilon^t \epsilon$

Prodotto interno o scalare di

$y-\hat{y}$ o di $\epsilon$

$D(\beta)=(y-\hat{y})^t (y-\hat{y}) = \epsilon^t \epsilon = minimo$

Criterio OLS:
$$min_{\beta}-D(\beta)= min_{\beta} \epsilon^t \epsilon$$
