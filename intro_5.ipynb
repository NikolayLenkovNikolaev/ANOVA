# Modello di regressione Lineare Multipla su Dati di popolazione

K var esplicative $X_1...X+k...X_k$ $k\geq 2$

- equazione del modello

$y= \beta_0 + \eta_1x_1 + ... \beta-jx_j +...+ \beta_kx_k + e = \beta_0 + \sum_{j=1}^k \beta_j x_j + e$


Distribuzione multipla di variabii $(Y, X_1...X_j...X_k)$


Sulla i-esima u.s. $u_i$ (k+1)- upla di valori

$(y_i,x_{i1}....x_{ij}...x_{ik}); o=1...N$

Equazione del modello espressa rispetto a $u_i$

$$y_i \\beta_0 + \sum_{j=1}^K \beta_j x_{ij} + e_i$$ i=1...N

$\hat{y_i} =\beta_0 + \sum_{j=1}^k \beta_j x_{ij} $ for all i

Problema: $I$ (k+1) parametri $\beta_j$ sono ignoti j=0,1...k

$\beta_{(k+1)*1} = [\beta_0, \beta_1....\beta_k]^t$

SOluzione: applicazione del criterio OLS

$min_{Î²}d(\beta) = min_{\beta} \sum_{i=1}^N (y_i - \hat{y_i})^2 =$

$= min_{\beta} \sum_{i=1}^N (y_i - \beta_0 - \sum_{j=1}^N \beta_j x_{ij}) = min_{\beta} \sum_{i=1}^N e_i^2$


dove:

- $e_i = y_i - \hat{y_i} = y_i - \beta_0 - \sum_{j=1}^k \beta_j x_{ij}$ for all i

residuo 0 errore piu l'u.s. $u_i$
