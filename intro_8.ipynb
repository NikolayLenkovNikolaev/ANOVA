1. prima di estrarre il campione
- var.dip.Y - si considerano le n v.c. $Y_i$ risultato alla i-esimo stazione (i=1..n); Y=(Y_1..Y_1...Y_n)

- var esplicative $X-j$ si assumono con manifestazione date o prefissate
- non hanno natira casuale (o aletoria)- sono deterministiche
- di fatto il modello e' costruito considerando le distribuzion condizionate di $Y_i|X_a...X_k; i=1...n$

- risultati campionari descritti dalle $n(k+1)- uple$

1.estraz: $Y_1,x_{11}, ...x_{1j}....x_{1k}$

....
i-esima estrazione: $Y_i,x_{i1, ...x_{ij}...x_{ik}}$

....

n-ma esttrazione - $(Y_n,x_{n1}...x_{nj}..x_{nk})$

n-estrazioni indipendenti

Equazione del modello
$$Y_i = \beta_0 + \beta_1 x_1 + ...+\beta_j x_{ij} +...+ \beta_k x_{ik} +\epsilon$$

$\epsilon_i$ - v.c. errore del modello i=1..n


$$Y_i = \beta_0 + \sum_{j=1}^k \beta_j x_{ij} +\epsilon_i$; i=1...n$

- parte deterministica - $\beta_0 + \sum_{j=1}^k \beta_j x_{ij}$
- parte casuale - $\epsilon$

Ipotesi sulle v.c. d'errore del modello:
- a: $E[\epsilon_i]= 0$ all i
- b1: $Var[\epsilon_i]=\sigma^2$ all i - ipotesi di omoschedasticiticita
- b2: $Cor(\epsilon, \epsilon_s)= 0$ $all- i\ne s$ - ipotesi di incorelazione
- b1 e b2 - sfericita degli errori

- c: $X_j$ a natura deterministica j=1...k con n>k+1 ; k<n+1

- d: $\epsilon \sim N(0, \sigma^2); i=1...n$

- $\epsilon_i$ - iid



Ipotesi specificate sulle v.c. $Y_i$

- destib condizionate: $Y_i|x_{i1}...x_{ij}..x_{ik} ; i=1..k$

- spesso si semplifica la notazione si una solo $Y_i$ intendendo camunque Y_i|x_{i1}...x_{ij}...x_{ik}$

- $Y_i$ e' una trasformazione lineare di $\epsilon_i$

a. $E[Y_i|x_{i1}...x_{ik}]= E[Y_i]=\mu_i= \beta_0 + \sum_{j=1}^k \beta_k x_{ij; i=1...n}$

b1. $Var[Y_i|x_{i1}...x_{ik}= Var[Y_i]= Var[\beta_0 + \sum_{j=1}^k \beta_j x_{ij}+ \epsilon_i]]= Var[\epsilon_i]= \sigma^2; all-i=1...n$

b2: $Cov[Y_i, Y_s]= Cov[\epsilon_i, \epsilon_s]= 0; all-i \ne s$

c: $X_j$ deterministiche all j con : k+1<n

d: $Y_i \sim N(\mu_i, \sigma^2; all - i 1..n)$

Osservazione: Y_i indipendente ma NON identicameente distribuite

$E[Y_i]=\mu_i \ne E[Y_s]=\mu_s$ par almeno una coppia (i,s)

$MST = \frac{SST}{n-1} = \frac{\sum_i (Y_i - \bar{Y})^2}{n-1}$ - varianza totale


$df(SSE)= n-(k+1)= n- rank(X)$ - si prendono k+1 df a causa della stima di $\beta$

$MSE = \frac{SSE}{n-(k+1)}= \frac{\sum_{i}(Y_i  - \hat{Y_i})^2}{n-(k+1)}= \xi^2$

- varianza residua ==$\xi^2$


$df(SSR)= df(SST) - df(SSE)= n-1-n + k + 1 = k = rank(X)- 1 =k+1-1$

$MSR= \frac{SSR}{k}= \frac{\sum_{i}(\hat{Y_i}- \bar{Y})^2}{k}$ - varianza spiegata

Indice di determinazione multipla:

$R^2 = \frac{SSR}{SST}= 1- \frac{SSE}{SST} \in [[0,1]$

- indice $R^2$ corretto (adjusted $R^2$) non e' un indice di bonta di adattamento. Non e' dunque normalizzato in [0,1]

$R_{adj}^2 = 1- \frac{\frac{  SSE  }{n-(k+1)   }}{\frac{SST}{n-1}   }= 1- \frac{MSE}{MST}$

confronto fra due moelli: si preferisce il modelo con valore di $R^2_{adj}$ piu elevato
