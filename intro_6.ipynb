Per inserire V nel modello come var.esplicat: CODIFICA BINARIA
- variabile indicatrice - dummy variable

Il modello di Y con V viene costruito utilizzando il regressione dummy X:
- $y = \beta_0 + \beta_1 + e$ o $y = \beta_0 + \beta_1 x_i + e_i; i=1...N$

dove:

$
\begin{cases}
    x=0 & \text{per $u_i = F$}\\
    x=1 & \text{per $u_i = M$}
  \end{cases}
$

F: $n_F $- numero di u.s in F
M: $n_M $- numero di u.s in M

$n_F = n_M = N$

l'equzione della realta intende due diverse definizioni:

$X= 0 (x_i = 0)$ per $u_i \in F; i=1...n_F$
- $y_i = \beta_0 + \beta_1 .0 + e_i = \beta_0 + e_i' i=1...n_F$

$X= 1 (x_i = 1)$ per $u_i \in M; i=1...n_M$
- $y_i = \beta_0 + \beta_1 .1 + e_i = \beta_0 + \beta_1 + e_i; i=1...n_M$

I parametri $\beta_0; \beta_1$ sono ignoti -> criterio OLS
- giustificato dal fatto che X e' una var. quantitativa discretta - variabile conteggio

$\hat{\beta_1} = \frac{Cov(X,Y)}{Var(X)}$

$\hat{\beta_0} = \mu_y - \hat{\beta_1}\mu_x$

- determinazione della formule OLS per $\beta_0$ e$\beta_1$ nel caso di una variabile qualitativa dicotomica

$\mu_x = \frac{1}{N} \sum_{i=1}^N x_i =\frac{1}{N}[(0+...+0)+ (1+...+1)]= $
- $()+..+0)= n_F / (1+...+1)= n_M$

$= \frac{1}{N} (0.n_F + 1.n_M)= \frac{n_M}{N}= p_M$

$p_M$: fraq.relativa o proporzionale di u.s. nel gruppo M



$Var[X]= \sigma_x^2 = \frac{1}{N} \sum_{i=1}^N (x_i - \mu_x)^2 = \frac{1}{N}[[(0-p_M)^2 + ...+ (0-p_M)^2] + [(1-p_m)^2 + ...+ (1-p_M)^2]]= $

$= \frac{1}{N} [(0-p_M)^2 n_F + (1-p_M)^2 n_M]$


Osser:

$p_F= \frac{n_F}{N}= \frac{N-n_M}{N}= 1- \frac{n_M}{N}= 1-p_M $

$p_M = 1-p_F$



$ Var[X]= \frac{1}{N}[p_M^2 n_F + (1-p_M)^2 n_M]= $

$= p_M^2 \frac{n_F}{N} + (1-p_M)^2 \frac{n_M}{N}$

$p_M^2 p_F^2 + (1-p_M)^2 p_M= $

$p_M^2 (1-p_M) + (1-p_M)^2 p_M= $

$p_M(1-p_M)[p_M + (1-p_M)]= $

$= p_M(1-p_M)= p_M.p_F$


$Cov(X,Y)= \frac{1}{N} \sum_i (x_i-\mu_x)(y_i -\mu_y)= $

$\frac{1}{N} \sum_i x_i y_i - \mu_x \mu_y= \frac{1}{N} \sum_i x_i y_y -p_M \mu_y$

- $M[XY]= \frac{1}{N} \sum_i x_i y_i = \frac{1}{N}[0.y_1 +...+ 0.y_{n_F}+1.y_{n_F +1}+...+ 1.y_N]= $

$\frac{1}{N} \sum_{i=1}^{n_M} y_i = \frac{1}{N}[\frac{1}{n_M} \sum_{i=1}^{n_M}y_i n_M]= $

$\frac{1}{N} \mu_y (M) n_M = \mu_y (M)\frac{n_M}{N} = \mu_y (M)p_M$

=> $Cor(X,Y) = \mu_y (M)p_M - \mu_y p_M$

$\mu_y = \frac{1}{N} \sum_{i=1}^N y_i = \frac{1}{N} [\sum_{i=1}^{n_F} y_i + \sum_{i=1}^{n_M} y_i]$

$= \frac{1}{N} [\frac{1}{n_F} \sum_{i=1}y_i n_F + \frac{1}{n_M} \sum_{i=1}y_i n_M$

$= \frac{1}{N}[\mu_y(F)n_F + \mu_y (M) n_M]= $

$\mu_y (F) \frac{n_F}{N} + \mu_y (M) \frac{n_M}{N}$

$= \mu_y (F) p_F + \mu_y (M) p_M$



$Cov(X,Y)= \mu_y (M) p_M -  p_M[\mu_y (F)p_F + \mu_y(M)p_M]= $

$\mu_y (M)p_M - \mu_y (F)p_F p_M - \my_y (M)p_M^2=$

$= \mu_y (M)p_M (1-p_M)- \mu_y (F)(1-p_M)p_M= $

$= p_M(1-p_M)[\mu_y(M)- \mu_y(F)]$



$\hat{\beta_0}= \mu_y - \hat{\beta_1}\mu_x = \mu_y - [\mu_y(M)-\mu_y(F)]p_M= $

$=\mu_y - \mu_y (M)p_M + \mu_y (F)p_M= $

$\mu_y (M)p_M + \mu_y (F)p_F - \mu_y (M)p_M+ \mu_y (F)p_M= $

$\mu_y (F)p_F + \mu_y (F)p_M= $

$\mu_y (F) (p_F + p_M) = \mu_y (F)$
- $p_F + p_M = 1$

Riepilogo soluzioni OLS:

$\hat{\beta_0}= \mu_y (F)$ - media di Y nel gruppo F

$\hat{beta_1}= \mu_y (M) - \mu_y (F)$ - differenza fra le medie di Y nel gruppo M e nel gruppo F


dove:

$\mu_y(F)= \frac{1}{n_F} \sum_{i=1}^{n_F}$

$\mu_y(M)= \frac{1}{n_M}= \frac{1}{n_M} \sum_{i=1}^{n_M}$

sono le medie condizionate di Y


Equazione del modello:

$\hat{y}= \hat{\beta_0} +\hat{\beta_1}x = \mu_y (F) + [\mu_y (M)- \mu_y (F)]x$

- se $x=0$ (v=F) $\hat{y}=\hat{\beta_0}=\mu_y (F)$

- se x=1 (x=M) $\hat{\beta_0} +\hat{\beta_1} = \mu_y(F) + [\mu_y(M)-\mu_y(F)]= \mu_y (M)$

### Parametri o coefficeinti standardizzati
- problema della confrontabilita fra entita degli effetti delle var.esplica (quantitative e/o qualitative) sulla var.dipendent Y in unmodello di regressione lineare multiple: "Quale regressione ha l'impatto piu forte su Y considerato l'insieme di tutti i regressori inseriti nel modello?"

- confontabilita fra entita degli effetti realizzati attraverso il ricorso ai parametri (coefficienti) standardizzati

Caso della rettaOLS:
- $Z_y = \frac{Y-\mu_y}{\sigma_y}; Z_x = \frac{Y-\mu_x}{\sigma_x}$

$\hat{Z_y}= \rho_{xy}Z_x$

$\rho_{xy}$ come Entita dell'effetto standardizzato di X su Y

Con K> 1 var.esplicative:
- confronto fra le entita degli effetti mediante il ricorso alle var standardizzte

es: $\hat{y} = 200 + 20000x_1 + 0.2x_2$

Anche se: $\hat{\beta_1} = 20000 > \hat{\beta_2}=0.2$ NON si puo concludee che $X_1$ ha un impattto su Y meggiore di $X_2$

Datp che $\hat{\beta_1}$ e $\hat{\beta_2}$ dipendono dalla diversa unita di misura delle var.esplicative e dalla loro diversa variabilita

Y in euro

$X_1$ in migliaia di euro

$X_2$ - in centesimi di euro

allora:

$\hat{\beta_1}= 20 000$ - ad ogni migliaio di euro in piu su $X_1$ corisponde un incremento medio su Y pari a 20 000 euro el netto di $X_2$

- $\hat{\beta_2}=0.2$ - ad ogni centesimo di euro in piu su $X_2$ corrispode n ncremento medio su Y pari a 0.2 euro al netto di $X_1$

Tuttavia

$X_1^` = \frac{X_1}{100 000} -> \hat{\beta_1}^` =\frac{\hat{\beta_1}}{100 000}=0.2$

- $\hat{\beta_1}^}`$ si interpreta come $\hat{\beta_2}$
- $X_1^`$ e $X_2$ hanno la stessa identica entita dell'effetto su Y


- confronto fra entita degli effettti sulla bse delle ENtita degli effetti standardizzati , anceh dette parametri o coefficienti standardizzati- standardized parameters, standardized Beta

$\hat{y} =\hat{\beta_0}+ \hat{\beta_1}x_1...+ \hat{\beta_k}.x_k$

- necessario costruire il modello con le var.standardizzate:

$Z_y = \frac{Y-\mu_y}{\sigma_y}; Z_j = \frac{Y-\mu_xj}{\sigma_xj}; j=1...K$

$\hat{Z_y} = \hat{\beta_0}^*Z_1 + \hat{\beta_0}^*.Z_2 + ...+ \hat{\beta_j}^*.Z_j + ...+ \hat{\beta_k}^*.Z_k$

- $\hat{\beta_j}^*$ - parametro standardizzto associato al refressore $X_j; j=1...K$

- $\hat{\beta_j}^*$  esprime l'entita dell'effetto standardizzato di $X_j$ su Y
- utilizzato peril confornto fra le entita dei K regressori insreriti nel modello

$\hat{\beta_j}^*= \hat{\beta_j} \frac{\sigma_{xj}}{\sigma_y}; j=1...K$

$\hat{\beta_j}^*$ - variazione media su Y standardizzata per ogni variazione in unita standardizzata di $X_j$ al netto delle altre var.standardizzate

- caso della retta OLS

$\hat{\beta_j}^*= \hat{\beta_1} \frac{\sigma_{x}}{\sigma_y} = \frac{\sigma_{xy}}{\sigma_x^2} \frac{\sigma_x}{\sigma_y}= \frac{\sigma_{xy}}{\sigma_x\sigma_y}= \rho_{xy}$


Oss: i coefficienti standardizzati $\hat{\beta_j}^*$ possono assumere tutti i valori reali e quindi anche valori negativi
- ai fini del confronto fra entita dei regressori si considera il loro valore assoluto

$\begin{equation}
|\hat{\beta_j}^*| = abs(\hat{\beta_j}^*) \begin{cases}
    \hat{\beta_j}^*, & \text{if $\hat{\beta_j}^* > 0 $}\\
    -\hat{\beta_j}^*, & \text{if $\hat{\beta_j}^* < 0 $}
  \end{cases}
\end{equation}$

Confronto: $X_r$ e' il regressione con la piu alta entia dell'effeto standardizzto se $|\hat{\beta_j}^*|= max_{j=1...K}$
