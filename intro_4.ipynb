# Studio della dipendenza in media
- riepilogo degli aspetti principali
1. Natura delle variabili
- Y var quantitativa dipendente
- X - var qualitativa o quantitativa indipendente

2. Struttura di dati di riferimento
- distribuzione di frequenze condizionate di $Y|x=x_j; j=1...J$
- $[(y_i, n_{ij}); i=1...I]; j=1...J$

- Manifgestazione $x_j$ di X interpretabili come "gruppi" rispetto ai queli le N u.s. sono suddivise
- ad ogni $x_j$ corisponde una vera e propria distribuzione di frequenza

3. Scelta del valore - rappresentante - di Y per ogni distribuzione condizionata di $Y|X= x_j$
- media aritmentica condizionata - $\mu_y (x_j)$ all j
- ragioni della scelta
3.1. proprieta di minimo della media aritmetica:
- data X var.quantitativa e $\mu_X= M[X]$ la sua media aritmetica, $\mu_x$ e' l'unco valore che minimizza la somma degli scarti al quadrato dei valori di X:
$\sum_{i=1}^N (x_i - A)^2 = minimo <=> A=\mu_X$

3.2. in un cotesto di modelistica adattata ai dati con il critertio OLS - quindi . anche X e' variabile quantitativa, le medie condizionate $\mu_y (x_j)$ si legano direttamente al concetto di REGRESSIONE
- X e Y variabili quantitative
- studio del elgame di dipendenza in media di Y da X mediante veri e propri modelli

- Modelli di Regressione

Problema: far corispondere ad ogni valore x di X un unico valore di Y nella distribuzione condizionata Y|X=x -> scelta di un valore "rapresentante" di Y|X=x- adeguato

- contesto di riferimento: criterio OLS

$\hat{y} = F(x)$ dove $f(.)$ tale che rende minimo l'errore (quadratico) nel sostituire $y$ con $\hat{y}$

- struttura di dati di partenza: $[(x_i, y_i, n_{ij}); i=1...I; j=1...J]$

-> struttura di dati obiettivo:
- $[(x_j, \hat{y_j}; n_{.j}); j=1..J]$

$\hat{y_j} = F(x_j); j=1...J$

$\sum_{i=1}^I \sum_{j=1}^J (y_i - \hat{y_j})^2 n_{ij} = \sum_{i=1}^I \sum_{j=1}^J (y_i - F(x_j))^2 n_{ij} = minimo$

- Soluzione: la funzione F e' la media condizionata di $Y|X=x$
- $\hat{y}= F(X)= \mu_y(x)$
- $\hat{y_j} = F(x_j)= \mu_y (x_j); j=1...J$

F- e' detta FUNZIONE DI REGRESSIONE

DImostrazione:
- coincide con al dimostrazione della proprieta di minimo della media aritmetica

$\sum_{i=1}^I \sum_{j=1}^J (y_i - \hat{y_j})^2 n_{ij} = minimo$

$\sum_{i=1}^I (y_i - \hat{y_j})^2 n_{ij}= minimo$ all j

$\sum_i (y_i - \hat{y_j})^2 n_{ij} = \sum_i [(y_i - \mu_y (x_j)) + (\mu_y (x_j) - \hat{y_i})]^2 n_{ij}$
- $-/+ \mu_y(x_j)$$

$\sum_i (y_i - \mu_y(x_j))^2 n_{ij} + \sum_i (\mu_y (x_j)- \hat{y_i})^2 n_{ij}$
- $(a+b)^2$

$= + 2(\mu_y(x_j)-\hat{y_i}) \sum_i (y_i - \mu_y (x_j))n_{ij}$
- $\sum_i (y_i - \mu_y (x_j))n_{ij} = 0$

$\sum_i (y_i - \mu_y (x_j))^2 n_{ij} + (\mu_y(x_j) -\hat{y_i})^2 \sum_i n_{ij}$
- $\sum_i n_{ij}$

$\sum_i (y_i - \mu_y (x_j))^2 n_{ij} + n_{.j}(\mu_y (x_j) - \hat{y_j})^2 = minimo$

- $\sum_i (y_i - \mu_y (x_j))^2 n_{ij} \geq 0 $
-  $ n_{.j}(\mu_y (x_j) - \hat{y_j})^2 \geq 0$


$<=> n_{.j} (\mu_y (x_j) - \hat{y_i})^2 = 0$

$<=> \hat{y_j} == \mu_y (x_j)$
- $\hat{y_j} = F(x_j); j=1...J$


Osservazione:

$|sum_i (y_i - \mu_y (x_j)) n_{ij} = n_{.j}[\frac{1}{n_{.j}} \sum_i (y_i - \mu_y (x_j))^2 n_{ij}]$

$n_{.j} \sigma_y^2 (x_j) => devianza di Y|X=x_j$

In conclusione ; come valore rappresenttnte di $Y|X=x$ si sceglie $\mu_y(x_j)$ poiche minimizza l'errore quadratico che si commete nel sostituire $y_i$ con $\mu_y (x_j)$ for all i,j



Definizione di Funzione di Regressione: e' una qualinque funzione F di X -(var quanitativa) tale che ad ogni valore $x_j$ osservato di X essocia la media condizionata $\mu_y (x_j)$ della distribuzione $Y|X=x_j$

$$\hat{y_j} = F(x_j) == \mu_y(x_j); j=1...J$$

- funzione di regressione come funione interpolante PER i punti di cordinate: $(x_j, \mu_y (x_j))$ all j

## Scomposizione della devianza totale di Y rispetto alla funzione di regressione
- richiamo : retta OLS

$SST= \sum_{i=1}^N (y_i - \mu_y)^2 = SSE + SSR= \sum_{i=1}^N (y_i - \hat{y_i})^2 + \sum_{i=1}^N (\hat{y_i}- \mu_y)^2$

dove: $\hat{y_i} =\hat{\beta_0} + \hat{\beta_1}x_i; i=1..N$

Funzione di regressione:
- $\hat{y} = F(x) = \mu_y (x)$

$SST = N\sigma_y^2= \sum_{i=1}^I (y_i - \mu_y)^2 n_{i.} = \sum_{j=1}^J \sigma_y^2 (x_j) n_{.j} + \sum_{j=1}^J [\mu_y (x_j) -\mu_y]^2 n_{.j}$


- $\sum_{j=1}^J \sigma_y^2 (x_j) n_{.j} = SSW= N\sigma^{*2}$
- $\sum_{j=1}^J [\mu_y (x_j) -\mu_y]^2 n_{.j} = SSB = N\bar{\sigma_y}^2$

- SSW - devianza NEI gruppo
- SSB - devianza FRA gruppi
- $SST = \sum_{i=1}^I (y_i - \mu_y)^2 n_{i.} = SSW + SSB$
    - $-/+ \hat{y_i}= \mu_y (x_j)$ all j

    

Contesto di modellistica:
- significato piu specifico per SSW e SSB
- $SSW = SSE_{RF}$ - devianza residua della funzione di regressione
- RF- regression funztion

- $SSB = SSR_{RF}$ - devianza spiegta dalla funzione di regressione

Indice di bonta di adattamento della funzione di regressione ai dati

$R_{RF}^2 =\frac{SSR_{RF}}{SSt} = \frac{SSB}{SST} = \eta_y^2$

Problema: l'impostazione di un sopre equivale all'impostazione del criterio WLS rispett a: $[(x_j, y_i, n_{ij}); i=1...I, j=1..J]$ ???

Risposta: si le due impostazioni sono equivalenti

Formalmente:
- $min_{\beta} D_{if}(\beta) = min_{\beta}\sum_{i=1}^I \sum_{j=1}^J (y_i - \hat{y_i}^2 n_{ij}) == min_{\beta} D_{MI}(\beta)= min_{\beta} \sum_{j=1}^J [\mu_y (x_j) - \hat{y_j}]^2 n_{.j}$
- IF - intepolant funztin
- MI - mean interpolant

dove: $\hat{y_j}= \hat{\mu_y} (x_j) = f(x_jl \beta)$j=1...J

Dim:
$D_{IF} (\beta) = \sum_i \sum_j (y_i - \hat{y_j})^2 n_{ij} = \sum_i \sum_j [(y_i - \mu_y (x_j)) + (\mu_y (x_j)- \hat{y_j})]^2 n_{ij}$
- $\mu_y(x_j)$

$\sum_i\sum_j (y_i -\mu_y (x_j))^2 n_{ij} ---A  + \sum_j (\mu_y (x_j)-\hat{y_j})^2 n_{.j} ---B$
- (a+b)^2

$= + 2\sum_i\sum_j (y_i -\mu_y (x_j))(\mu_y (x_j) -\hat{y_j})n_{ij} ---C$


$C: 2\sum_j (\mu_y (x_j) -\hat{y_i}) \sum_i (y_i - \mu_y (x_j))n_{ij}$
- $\sum_i (y_i - \mu_y (x_j))n_{ij} =0$ all j

$D_{IF} (\beta)= \sum_i \sum_j (y_i - \hat{j})^2 n_{ij} =$

$= \sum_i \sum_j [y_i - \mu_y (x_j)]^2 n_{ij} + \sum_j [\mu_y (x_j) -\hat{y_i}]^2 n_{.j}$

Semplificato di A e B:

$\sum_{j=1}^J \sigma_y^2 (x_j)n_{.j} =SSW = SSE_{RF} = N\sigma_y^{*2}$
