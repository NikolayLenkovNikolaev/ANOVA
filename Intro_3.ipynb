- analisi della dipendenza in Media di Y da X
- def: di Indipendenza in Media (I.M) di Y da X

$\mu_y(x_1)= \mu_y(x_2)=...\mu_y(x_j)= ...\mu_y(x_J)=\mu_y$

- X e' I.M da Y <=> $\mu_x(y_1)= ....= \u_x(y_i)= ....=\mu_x(y_I)= \mu_x$

Def di DIPENDENZA IN MEDIA di Y da X:

- $\mu_y(x_j) \ne \mu_y(x_r)$ per almeno un (j,r) con $j\ne r$

- $\mu_y (x_j) \ne \mu_y$

OSS:

se Y e' IM da X:

$\bar{\sigma_y}^2 = Var[M(Y|X)] = \frac{1}{N} \sum_j [\mu_y (x_j) - \mu_y]^2 n_{.j} = $
- $\mu_y (x_j) =\mu_y$

$\frac{1}{N} \sum_j (\mu_y - \mu_y)^2 m_{.j}= 0$

=> $\sigma_y^2 = \sigma_y^{*2} + \bar{\sigma_y}^2 == \sigma_y^{*2} = M[Var[Y|X]]$



RAPPORTO DI CORRELAZIONE DI PEARSON:

$\eta_y^2 = \eta_{y|x}^2 = \frac{\bar{\sigma_y}^2}{\sigma_y^2} = \frac{\sigma_y^2 - \sigma_y^{*2}}{\sigma_y^2} = 1- \frac{\sigma_y^{*2}}{\sigma_y^2}$

Caratteristiche di $\eta_y^2$

1. $\eta_y^2 \in [0,1]$ - indice normalizzato
2. significato degli estremi dell'intervallo di definizoone:

$\eta_y^2 = 0 <=> \bar{\sigma_y}^2=0 <=> \sigma_y^2= \sigma_y^{*2}$ <=> Y e\ IM da X

$\eta_y^2 = 1 <=> \bar{\sigma_y}^2=\sigma_y^2 <=> \sigma_y^{*2}=0$ <=> perfetta dipendenza inmedia di Y da X
